```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

# Same task using sklearn built-in linear regression model: predict revenue for next year

```python
def make_data(n_samples=100):
    rnd = np.random.RandomState(42)
    x = rnd.uniform(1, 7, size=n_samples)
    y_no_noise = (np.sin(4 * x) + x + 5)
    y = (y_no_noise + rnd.normal(size=len(x))) / 2
    y *= 5
    return x.reshape(-1, 1), y
```

```python
# create data
X, y = make_data()
```

```python
plt.scatter(X, y)
plt.title("Scatter Plot")
plt.xlabel("Advertising Expediture (million dollar)")
plt.ylabel("Sales Revenue (million dollar)");
```

```python
# create model object
reg_model = LinearRegression()
```

```python
# train model on all available data
reg_model.fit(X, y)
```

```python
# print learned parameters
intercept = reg_model.intercept_
coef = reg_model.coef_
print(intercept)
print(coef)
```

```python
# using model to predict
reg_model.predict([[5.2]])
```

#Ex1: Train test split
Using 80% of data as training data to train a linear regression model, then test the model performance on the remaining 20% data, using RMSE metric.

```python
# Step 0. import needed packages or functions
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Step 1. using train_test_split function to split data into training and test set
# test_size =0.2, random_state=1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)
```

```python
# Step 2. Creating a linear regression model
reg_1 = LinearRegression()
```

```python
# Step 3. using fit funciton to train the model on training data
reg_1.fit(X_train, y_train)
```

```python
# Step 4. using the learned model to get predictions on test data
y_pred = reg_1.predict(X_test)
```

```python
# step 5. print model's intercept term and coefficient.
```

```python
intercept = reg_1.intercept_
coef = reg_1.coef_
print(intercept)
print(coef)
```

```python
mean_squared_error(y_test, y_pred)
```
