```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage import data
from scipy.signal import correlate2d
```

```python
simple_input = np.array([[ 2,  8,  6,  9,  1],
 [ 4,  5,  0,  3,  7],
 [ 6,  1,  9,  2,  4],
 [ 0,  3,  8,  7,  5],
 [ 9,  6,  4,  1,  0]])
```

```python
# Define the 3x3 filter for detecting horizontal edges
kernel = np.array([[1, 1, 1],
                   [0, 0, 0],
                   [-1, -1, -1]])
```

```python
# Apply the convolution operation
convolved_simple_input = correlate2d(simple_input, kernel, mode='valid')
```

```python
convolved_simple_input
```

```python
# Print the original image and the convolved image
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(simple_input, cmap='gray')
axes[0].set_title("Original Image")

axes[1].imshow(convolved_simple_input, cmap='gray')
axes[1].set_title("Convolved Image")

plt.show()
```

## Now using the kernel on an actual image

```python
image = data.camera()
```

```python
# using the conv function from cv2 library
convolved_image = cv2.filter2D(image, -1, kernel)
```

```python
# Print the original image and the convolved image
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")

axes[1].imshow(convolved_image, cmap='gray')
axes[1].set_title("Convolved Image (Horizontal Edge Filter)")

plt.show()
```

```python
# Define the 3x3 filter for detecting vertical edges
kernel_2 = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])
```

```python
convolved_image_2 = cv2.filter2D(image, -1, kernel_2)
```

```python
# Print the original image and the convolved image
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")

axes[1].imshow(convolved_image_2, cmap='gray')
axes[1].set_title("Convolved Image (Vertical Edge Filter)")

plt.show()
```
